```python
%matplotlib inline
import pandas
from sklearn import linear_model as lm
from plotnine import *

root = '/Users/rix0rrr/Google Drive/Hackathon 28_4'
filename = root + '/Team 3/Input voor de lineare regressie-v4.csv'


df = pandas.read_csv(filename, index_col='BEVOEGD_GEZAGNAAM', encoding='utf-8')

catted = df.loc[df['JAAR'] == 2016].sort_values('AANDEEL_FORMATIE_OP_LB')
catted.loc[:, 'cat'] = 'overig'
catted.loc[:50, 'cat'] = 'slechtst'
catted.loc[-50:, 'cat'] = 'best'
catted = catted[['cat']]

y2016 = df.loc[df['JAAR'] == 2016, ['AANTAL_LEERLINGEN', 'AANDEEL_FORMATIE_OP_LB']]
y2013 = df.loc[df['JAAR'] == 2012, ['AANTAL_LEERLINGEN', 'AANDEEL_FORMATIE_OP_LB']]

Y = y2013.join(y2016, rsuffix='2016').join(catted)
Y['LEERLING_GROEI'] = (Y['AANTAL_LEERLINGEN2016'] - Y['AANTAL_LEERLINGEN'])/Y['AANTAL_LEERLINGEN']
Y['LB_GROEI'] = Y['AANDEEL_FORMATIE_OP_LB2016'] - Y['AANDEEL_FORMATIE_OP_LB']

(ggplot(Y, aes('LEERLING_GROEI', 'LB_GROEI', color='factor(cat)'))
 + geom_point()
 + stat_smooth(method='lm'))
```
